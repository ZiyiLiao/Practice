# Scrapping Web Text Data

```python
import requests
import numpy as np
import pandas as pd
import re
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
url = "https://www.espn.com/nba/team/schedule/_/name/BKN/seasontype/2"
r = requests.get(url)
nets1819 = r.text
```

```python
len(nets1819)
```

```python
date_expr = '[A-Z][a-z]{2},\s[A-Z][a-z]{2}\s\d+'
date = re.findall(date_expr, nets1819)
len(date)
```

```python
away_expr = 'class=\"pr2\">(?:vs|@)</span>'
away = re.findall(away_expr, nets1819)
len(away)
```

```python
home = list()
for element in away:
    if "@" in element:
        home.append(0)
    else:
        home.append(1)
```

```python
opponent_expr = '(?:[a-z]|-|[0-9])+\"><img'
opponent = re.findall(opponent_expr, nets1819)
len(opponent)    
```

```python
name = list()
for element in opponent:
    name.append(element.rstrip('\"><img'))
name[0:10]
```

```python
schedule = pd.DataFrame(
    {'date': date,
     'opponent': name,
     'home': home
    })
schedule
```
